# Question - 1

### Part - A

```python

import sympy as sp

# Declaring symbolic variable
x, y = sp.symbols('x y')

# Defining function f(x,y) = x^2 + y^2 + x*(y+2) + cos(3x)
f = x**2 + y**2 + x*(y + 2) + sp.cos(3*x)

# Computing partial derivative of f with respect to x (∂f/∂x)
df_dx = sp.diff(f, x)

print("\n Derivative with respect to x, ∂f/∂x =", df_dx)
```

### Part - B

```python

# Computing partial derivative of f with respect to y (∂f/∂y)
df_dy = sp.diff(f, y)

print("\n Derivative with respect to y, ∂f/∂y =", df_dy)
```

### Part - C

```python

import numpy as np
import matplotlib.pyplot as plt

# Defining function
def f(x, y):
    return x**2 + y**2 + x*(y + 2) + np.cos(3*x)

# Defining function gradients
def grad(x, y):
    df_dx = 2*x + y + 2 - 3*np.sin(3*x)
    df_dy = 2*y + x
    return np.array([df_dx, df_dy])

# Gradient descent parameters
lr = 0.1  
max_iters = 1000
tol = 1e-6

# Intial Random grid point guess
x, y = 1.0, 1.0  


# Running GD and storing descent path
path = [(x, y)]
for i in range(max_iters):
    dx, dy = grad(x, y)
    x_new, y_new = x - lr*dx, y - lr*dy
    path.append((x_new, y_new))
    if np.linalg.norm([x_new - x, y_new - y]) < tol:
        x, y = x_new, y_new
        print(f"\n Converged in {i+1} iterations.")
        break
    x, y = x_new, y_new
else:
    print(f"\n Reached max iterations without full convergence.")


# Result display
min_f = f(x, y)
print(f"\n Approximate minimum at x = {x:.6f}, y = {y:.6f}")
print(f"\n Function value at minimum f(x,y) = {min_f:.6f}")
```

### Part - D

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Path from previous cell
path = np.array(path)
xs, ys = path[:,0], path[:,1]
zs = f(xs, ys)

# Create meshgrid for the surface plot
x_range = np.linspace(min(xs)-0.5, max(xs)+0.5, 100)
y_range = np.linspace(min(ys)-0.5, max(ys)+0.5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)  # Compute function values for the surface

# Figure plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(
    X, Y, Z,
    rstride=5, cstride=5,
    alpha=0.7, edgecolor='none'
)

# GD path 
ax.plot(xs, ys, zs, color='r', marker='o', linewidth=2, label='GD path')
ax.scatter(xs[0], ys[0], zs[0], color='g', s=50, label='Start')
ax.scatter(xs[-1], ys[-1], zs[-1], color='k', s=50, label='End')

# Labels and legend
ax.set_title('Gradient Descent Path of f(x,y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.tight_layout()
plt.show()
```
